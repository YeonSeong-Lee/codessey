```python
from email.contentmanager import raw_data_manager

import pandas as pd

# Load the dataset
df = pd.read_csv('./data/prob-0101.csv')
df.head()
```

```python
import plotly.express as px

# Group by release date and count movies
movie_counts = df['개봉일'].value_counts().reset_index()
movie_counts.columns = ['개봉일', '영화수']
movie_counts = movie_counts.sort_values('개봉일')

# Plot interactive bar chart
fig = px.bar(movie_counts, x='개봉일', y='영화수', title='개봉일별 영화 수', labels={'개봉일': '개봉일', '영화수': '영화 수'}, hover_data={'영화수': True})
fig.update_layout(xaxis=dict(title='개봉일', tickangle=45), yaxis_title='영화 수')
fig.show()
```

```python
unique_values = {
    '배급사': df['배급사'].unique(),
    '감독': df['감독'].unique(),
    '출연진': df['출연진'].unique(),
    '장르': df['장르'].unique()
}
```

```python
unique_actor = sorted(set(actor for sublist in df['출연진'] for actor in sublist))
unique_actor
```

```python
import random

def get_user_actor_choice(re_randomize=False):
    if re_randomize:
        random.shuffle(unique_actor)
    random_actors = unique_actor[:3]
    user_input = input(f"출연진 중 한 명을 선택하세요: {random_actors}, 원하는 배우가 없다면 `re`를 입력하세요.")
    if user_input == 're':
        return get_user_actor_choice(True)
    if user_input in random_actors:
        return user_input
    else:
        return get_user_actor_choice(False)
```

```python
selected_actor = get_user_actor_choice().strip()
recommended_movies = df[df['출연진'].apply(lambda cast: selected_actor in cast)]
recommended_movies
```
